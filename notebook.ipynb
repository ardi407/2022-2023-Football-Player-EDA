````python
# Fungsi untuk menyimpan kolom yang dibutuhkan saja
def delete_column(dataframe):
    '''
    Akan menghapus kolom jika nama kolom dari sebuah dataframe jika kolom tersebut tidak ada dalam list di bawah!
    '''
    columns_needed = [
        'Player', 'Nation', 'Pos', 'Squad', 'Comp', 'Age', '90s',
        'Goals', 'Shots', 'SoT%', 'G/Sh', 'G/SoT', 'Assists', 'GcaDrib',
        'Int', 'Tkl', 'TklW', 'Clr', 'Blocks', 'AerWon%',
        'PasTotCmp%', 'PasShoCmp%', 'PasMedCmp%', 'PasLonCmp%', 'PasAss',
        'SCA', 'GCA', 'ScaDrib',
        'AerWon%', 'Min', 'MP']
    
    # untuk output
    new_df = pd.DataFrame()

    # Loop kolom
    for column in dataframe.columns:
        if column not in columns_needed:
            dataframe.drop(column, axis=1)
        else:
            new_df[column] = dataframe[column]
    
    # membuat kolom baru
    new_df['AvgMinMatch'] = new_df['Min']/new_df['MP']
    return new_df

# Memanggil fungsi di atas dan simpan kedalam variable dataframe baru
cleaned_df = delete_column(df)
cleaned_df.head()
````

*Membagi table menjadi beberapa tabel berdasarkan kategori data pribadi pemain dan aspek skills*

````python
# Membuat table berdasarkan kategori
# Table data pribadi pemain
player = ['Player', 'Nation', 'Pos', '90s', 'Squad', 'Comp', 'Age', 'Goals']
offensive = ['Player', 'Pos', '90s', 'Goals', 'Shots', 'SoT%']
defensive = ['Player', 'Pos', '90s', 'Int', 'Tkl', 'TklW', 'Clr', 'Blocks', 'AerWon%']
passing = ['Player',  'Pos', '90s','PasTotCmp%', 'PasShoCmp%', 'PasMedCmp%', 'PasLonCmp%', 'PasAss']
playmaking = ['Player', 'Pos', '90s','SCA', 'GCA', 'ScaDrib']
physical = ['Player', 'Pos','90s','AerWon%', 'Min', 'MP','AvgMinMatch']

# Fungsi untuk mengenerate table
def generate(dataframe,columns):
    new_df = pd.DataFrame()
    for column in columns:
        new_df[column] = dataframe[column]
    return new_df

#  generate table 
players = generate(cleaned_df, player)
offensives = generate(cleaned_df, offensive)
defensives = generate(cleaned_df, defensive)
passings = generate(cleaned_df, passing)
playmakings = generate(cleaned_df, playmaking)
physicals = generate(cleaned_df, physical)
````

##### 1. Menentukan dan Menghitung Overall Rating

- Overall rating setiap skill direpresentasikan dengan skala 0-100
- Perhitungan overall rating setiap skill dihitung dengan cara yang berbeda-beda. 

*a. Offensives Rate*

Dengan asumsi Shots adalah jumlah tembakan per laga, menghitung rating offensive skill dilakukan dengan cara:

````python
# goal per match goals dibagi jumlah menit/90 atau 90s
goal_per_match = offensives['Goals'] / offensives['90s']

# Ubah ke dalam persentase
goal_per_match_rating = goal_per_match * 100

# shots total diasumsikan sbg jumlah tembakan per laga, jadi dikali 90s
shots_total = offensives['Shots'] * offensives['90s']

# untuk perbandingan dengan value tertinggi dari pemain lain
max_shots_total = shots_total.max()
shot_per_match_rating = shots_total/max_shots_total * 100 # dibuat persentase

# ssama halnya dengan shots
SoT = offensives['SoT%'] * offensives['90s']
max_Sot = SoT.max()

#persentase
SoT_rate = SoT/max_Sot * 100

# menghitung offensive rate dengan formula sum/3 (average dari kolom2 yang sudah dihitung)
offensives['Offensive Rate'] = (goal_per_match_rating + shot_per_match_rating + SoT_rate) / 3
offensives['Offensive Rate'] = offensives['Offensive Rate']/offensives['Offensive Rate'].max()*100
offensives.sort_values('Offensive Rate', ascending=False).head()
````

*b. Defensives Rate*

Dengan asumsi Int, TklW, Clr, dan Blocks adalah angka per match, rating defensives tables, dilakukan dengan formula: <br>

````python
# asumsi semua data adalah jumlah per laga kecuali aerWon

int = defensives['Int'] *defensives['90s']
tkl = defensives['Tkl'] *defensives['90s']
tklW = defensives['TklW'] *defensives['90s']
Clr = defensives['Clr'] *defensives['90s']
Blocks = defensives['Blocks'] *defensives['90s']

# jumlahkan hasil di atas
sum_90s = (int + tkl + tklW + Clr + Blocks)
max_sum_90s = sum_90s.max()

# sum/max dari value yang ada dikali 100 (persentase)
sum_rating = (sum_90s/max_sum_90s) * 100

# dipisah karena dalam bentuk %
aer_won = defensives['AerWon%'] * defensives['90s']
max_aer_won= aer_won.max()
aer_won_rate = aer_won/max_aer_won * 100

# hitung defensive rate
defensives['Defensive Rate'] = (sum_rating + aer_won_rate)/2
defensives['Defensive Rate'] = defensives['Defensive Rate']/defensives['Defensive Rate'].max()*100
defensives.sort_values('Defensive Rate', ascending=False).head(10)
````

*c. Passing Rate*

Dengan sumsi PassAss adalah angka per match, perhitungan passing rating dilakukan dengan cara:

````python
# hitung masing-masing kolom 
passtotal= passings['PasTotCmp%'] * passings['90s']
passcomp = passings['PasShoCmp%'] * passings['90s']
passmed = passings['PasMedCmp%'] * passings['90s']
passL = passings['PasLonCmp%']	* passings['90s']

# jumlahkan hasil di atas
sum_pass = passtotal + passcomp + passmed + passL
max_pass = sum_pass.max()
pass_rate = sum_pass/max_pass * 100 # ubah kedalam persentase

# Ini dipisah dengan value-value di atas karena memiliki skala yanng berbda
passAssist = passings['PasAss'] * passings['90s']
max_passAssist = passAssist.max()
passAssist_rating = (passAssist/max_passAssist) * 100

# hitung value passing rate
passings['Passing Rate'] = (pass_rate + passAssist_rating)/2
passings['Passing Rate'] = passings['Passing Rate']/passings['Passing Rate'].max()*100
passings.sort_values('Passing Rate', ascending=False).head()
````

*d. Playmakings Rate*

Dengan asumsi SCA, GCA, dan Sca Drib adalah angka per match, menghitung playmakings rating dilakukan dengan cara:

````python
# hitung masing2 kolom
sca = playmakings['SCA'] * playmakings['90s']
gca = playmakings['GCA'] * playmakings['90s']
sca_drib = playmakings['ScaDrib'] * playmakings['90s']

# hitung persentase
sca_rate = sca/sca.max() * 100
gca_rate = gca/gca.max() * 100
sca_drib_rate = sca_drib/sca_drib.max() * 100

# hitung kolom playmaking rate
playmakings['Playmaking Rate'] = (sca_rate + gca_rate + sca_drib_rate)/3 
playmakings['Playmaking Rate'] = playmakings['Playmaking Rate']/playmakings['Playmaking Rate'].max()*100 
playmakings.sort_values('Playmaking Rate', ascending=False).head()
````

*e. Physicals Rate*

Menghitung overall rating pyshical ability dilakukan dengan cara:

````python
# hitung kolom-kolom yang dibutuhkan untuk rate
avg_p_match = physicals['AvgMinMatch']/90 * 100
aer_won_rate = physicals['AerWon%']

# hitung hasil physical rate
physicals['Physical Rate'] = (aer_won_rate + avg_p_match)/2
physicals['Physical Rate'] = physicals['Physical Rate']/physicals['Physical Rate'].max()*100
physicals.sort_values('Physical Rate',ascending=False).head()
````

**Player Overall Rating**

Menggabungkan seluruh rating pemain ke dalam satu table

````python
# Tambahkan semua kolom yang dibutuhkan yang rating yang telah dibuat ke dalam sebuah table
overall_rating = pd.DataFrame()
overall_rating = players.copy()
````

````python
def add_columns(dataframe1,dataframe2, rate):
    '''
    fungsi mengambil 3 perimeter yaitu, dataframe 1 dengan yang lainnya, serta rate untuk pengambilan kolom menggunakan pd.merge
    '''
    merged = pd.DataFrame()
    merged = pd.merge(left=dataframe1, right=dataframe2[['Player',rate]], on='Player', how='left')

    return merged

# panggil fungsi di atas
overall_rating = add_columns(overall_rating, offensives, 'Offensive Rate')
overall_rating = add_columns(overall_rating, defensives, 'Defensive Rate')
overall_rating = add_columns(overall_rating, passings, 'Passing Rate')
overall_rating = add_columns(overall_rating, playmakings, 'Playmaking Rate')
overall_rating = add_columns(overall_rating, physicals, 'Physical Rate')
overall_rating.head()
````

Menghitung Overall Rating Pemain

Untuk membuat penilaian lebih sesuai dengan posisi setiap pemain, overall rating didapatkan dengan cara memberikan bobot untuk masing-masing skills seperti berikut:

````python
# Mengatur bobot rates sesuai dengan posisi pemain
bobot_rating = {'FW': {'Offensive Rate': 0.65,
                       'Playmaking Rate': 0.1,
                       'Passing Rate':0.1,
                       'Defensive Rate': 0.05,
                       'Physical Rate': 0.1},
                'MF' : {'Offensive Rate': 0.1,
                       'Playmaking Rate': 0.5,
                       'Passing Rate':0.2,
                       'Defensive Rate': 0.1,
                       'Physical Rate': 0.1},
                'DF': {'Offensive Rate': 0.1,
                       'Playmaking Rate': 0.1,
                       'Passing Rate':0.1,
                       'Defensive Rate': 0.4,
                       'Physical Rate': 0.3},
                }
````

````python
def calculate_overall_rate(dataframe, bobot):
    '''
    fungsi mengambil sebuah dataframe dan bobot penilaian
    '''
    result_df = dataframe.copy() # copy dataframe
    for posisi, value in bobot.items():
        filter = dataframe['Pos'] == posisi # mengatur filter untuk kondisi loc
        for aspek, rate in value.items():
            # melakukan perhitungan terhadap cell yang tepat sesuai dengan loc, artinya perhitungan dilakukan bersamaan, tidak looping ke isi dataframe
            result_df.loc[filter, aspek] = result_df.loc[filter, aspek] * rate
    
    # menjumlahkan hasil perkalian masing-masing rate dari result_df
    dataframe['Overall Rating'] = result_df[['Offensive Rate', 'Defensive Rate', 'Passing Rate', 'Playmaking Rate', 'Physical Rate']].sum(axis=1)

    # Menghitung overall rating berdasarkan persentase dengan max value, dikalikan 95 untuk membuat standard (agar tidak ada pemain dengan nilai 100 overall rating).
    dataframe['Overall Rating'] = dataframe['Overall Rating']/dataframe['Overall Rating'].max()*95
    
    return dataframe
````

````python
overall_rating = calculate_overall_rate(overall_rating, bobot_rating)
# hasil pemanggilan fungsi
overall_rating.sort_values('Overall Rating', ascending=False).head(8)
````

````python
# Import modules yang dibutuhkan
from matplotlib import pyplot as plt
import seaborn as sns 
from scipy import stats
````

Fungsi untuk menarik data, mean, median, mode, std, skew, dan kurtosis

````python
def central_tendency(column):
    '''
    fungsi yang menggambil sebuah series dari dataframe
    '''

    # hiutung central tendency data
    mean = column.mean()
    median = column.median()
    mode = column.mode().iloc[0]
    std = column.std()
    skew = column.skew()
    kurtosis = column.kurtosis()

    # print text yang mengandung hasil perhitungan
    print(f"Kolom {column.name} memiliki central tendency data sebagai berikut:")
    print(f"Rata-rata = {mean:.2f}")
    print(f"Median = {median:.2f}")
    print(f"Modus = {mode}")
    print(f"Standard Deviation = {std:.2f}")
    print(f"Nilai skewness = {skew:.2f}")
    if skew < -1:
        print("Karena nilai skew di bawah negative 1, maka bisa dikatakan bahwa data berupa berbentuk negatif skew yang ekstrim")
    elif skew > 1:
        print("Karena nilai skew di atas 1, maka bisa dikatakan bahwa data berupa berbentuk positif skew yang ekstrim")
    elif skew < -0.5:
        print("Karena nilai skew lebih rendah dari negatif 0.5 namun lebih tinggi dari negatif 1, maka bisa dikatakan bahwa data berupa berbentuk negative skew menengah")
    elif skew > 0.5:
        print("Karena nilai skew lebih tinggi dari 0.5 namun lebih rendah dari 1 , maka bisa dikatakan bahwa data berupa berbentuk positif skew menengah")
    else:
        print("Nilai skewness mendekati nol, maka data hampir memiliki sifat simetris")
    print(f"Nilai Kurtosis = {kurtosis:.5f}")
    
    if kurtosis == 0.00:
        print("Data memiliki distribusi mesokurtic atau distribusi bersifat normal")
    elif kurtosis < 0:
        print("Data memiliki sifat platykurtic atau distribusi data tidak normal")
    elif kurtosis > 0:
        print("Data memiliki sifat leptokurtic atau distribusi data tidak normal")

    # set up chart dengan pandas
    fig = plt.figure(figsize=(5, 4))
    column.plot(kind='density')
    plt.xlabel(column.name)
    plt.ylabel('denisty')
    plt.title('Density Chart of '+ column.name)
    plt.axvline(column.mean(), color='black', linestyle='dashed', linewidth=1, label='Mean')
    plt.axvline(column.median(), color='red', linestyle='dashed', linewidth=1, label='Median')
    plt.axvline(column.std(), color='purple', linestyle='dashed', label = 'Std')
    plt.legend()
    plt.show()

    # tidak perlu mengembalikan apa2. Fungsi hanya perlu dipanggil
    return None
````

**Korelasi antara usia pemain dan overall rating**

Overall rating diketahui tidak memiliki distribusi yang normal, sehingga akan dilakukan spearman dan kendall untuk mengetahui adakah hubungan antara usia dan overall rating pemain. Uji hypothesis dengan significanec level 0.05 akan dilakukan

Hypotheses
> Null Hypothesis: "Tidak terdapat hubungan yang signifikan antara overall rating dan usia pemain"

> Alternative Hypothesis: "Terdapat hubungan yang signifikan antara overall rating dan usia pemain"

````python
# menampilkan lmplot untuk melihat visualisasi data
sns.lmplot(overall_rating, x='Overall Rating', y='Age')
plt.show()
````

````python
# menghitung corr rho dan tau, p value dari test spearman dan kendall
corr_rho, p_val_s = stats.spearmanr(overall_rating['Overall Rating'],overall_rating['Age'])
corr_tau, p_val_k = stats.kendalltau(overall_rating['Overall Rating'], overall_rating['Age'])

print(f"rho-correlation: {corr_rho:.2f}, p-value: {p_val_s}")
print(f"tau-correlation: {corr_tau:.2f}, p-value: {p_val_k}")
````

> Dari nilai rho dan tau, terlihat adanya hubungan negatif yang lemah. <br>
> p_value dari kedua test sama-sama lebih kecil dari significance level (0.05), artinya H1 diterima yaitu: Terdapat hubungan yang signifikan antara overall rating dan usia pemain.

> Hal ini mengindikasikan bahwa pemain dengan usia lebih muda cenderung memiliki overall rating yang lebih baik.

##### 3. Statistik Inferensial

Pada bagian ini, akan dilakukan proses statistik inferensial untuk menjawab pertanyaan berikut:<br>
**Bagaimankah performa pemain dalam mencetak gol berdasarkan posisi? Adakah hubungan antara posisi dengan jumlah gol yang dicetak setiap pemain?**

Uji Anova akan digunakan untuk menjawab pertanyaan tersebut dengan menggunakan significance level = 0.05 dan Hypotheses:


> H0 = 'Tidak terdapat perbedaan yang signifikan dalam jumlah goal dan posisi pemain' <br>
> H1 = 'Terdapat perbedaan yang signifikan dalam jumlah goal dan posisi pemain'

````python
# ekstrak data berdasarkan posisi
FW = overall_rating['Goals'][overall_rating['Pos']=='FW']
MF = overall_rating['Goals'][overall_rating['Pos']=='MF']
DF = overall_rating['Goals'][overall_rating['Pos']=='DF']
print(f'Rata-rata Goal posisi FW: {FW.mean():.2f}')
print(f'Rata-rata Goal posisi MF: {MF.mean():.2f}')
print(f'Rata-rata Goal posisi DF: {DF.mean():.2f}')
````

>terlihat perbedaan rata-rata antar posisi.

````python
# menghitung p value menggunakan stats.f_oneway (ANOVA)
result = stats.f_oneway(FW, MF, DF)
print('P-value:',result.pvalue)
````

P-Value jauh lebih kecil dari significance level (0.05)<br> H1 diterima: **Terdapat perbedaan yang signifikan dalam jumlah goal dan posisi pemain**<br>
Hal ini mengindikasikan bahwa posisi pemain sangat mempengaruhi jumlah goal yang dicetak.
